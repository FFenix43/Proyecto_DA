```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Leer el archivo CSV
archivo1 = 'C:/Users/Usuario/Downloads/Internet_Penetracion.csv'
df1 = pd.read_csv(archivo1)

# Reemplazar las comas por puntos en los valores de la columna "Accesos por cada 100 hogares"
df1['Accesos por cada 100 hogares'] = df1['Accesos por cada 100 hogares'].str.replace(',', '.')

# Convertir la columna "Accesos por cada 100 hogares" a tipo numérico
df1['Accesos por cada 100 hogares'] = pd.to_numeric(df1['Accesos por cada 100 hogares'], errors='coerce')

# Crear un gráfico de barras agrupadas por provincia y trimestre
plt.figure(figsize=(12, 6))
sns.barplot(data=df1, x='Provincia', y='Accesos por cada 100 hogares', hue='Trimestre')
plt.title('Accesos por cada 100 hogares por Provincia y Trimestre')
plt.xlabel('Provincia')
plt.ylabel('Accesos por cada 100 hogares')
plt.xticks(rotation=45)
plt.legend(title='Trimestre')
plt.show()

```

Los mayores accesos de internet se dieron en el tercer trimestre mientras que en en los trimestres 1 y 4 fueron los mas bajos. 

```python
# Leer el archivo CSV
archivo1 = 'C:/Users/Usuario/Downloads/Internet_Penetracion.csv'
df1 = pd.read_csv(archivo1)

# Reemplazar las comas por puntos en los valores de la columna "Accesos por cada 100 hogares"
df1['Accesos por cada 100 hogares'] = df1['Accesos por cada 100 hogares'].str.replace(',', '.')

# Convertir la columna "Accesos por cada 100 hogares" a tipo numérico
df1['Accesos por cada 100 hogares'] = pd.to_numeric(df1['Accesos por cada 100 hogares'], errors='coerce')

# Crear un gráfico de barras agrupadas por provincia y año
plt.figure(figsize=(12, 6))
sns.barplot(data=df1, x='Provincia', y='Accesos por cada 100 hogares', hue='Año')
plt.title('Accesos por cada 100 hogares por Provincia y Año')
plt.xlabel('Provincia')
plt.ylabel('Accesos por cada 100 hogares')
plt.xticks(rotation=45)
plt.legend(title='Año')
plt.show()
```

```python
# Leer el archivo CSV
archivo2 = 'C:/Users/Usuario/Downloads/Internet_BAF.csv'
df2 = pd.read_csv(archivo2)

# Reemplazar los puntos y comas por nada en los valores de las columnas numéricas
columns_to_replace = ['Banda ancha fija', 'Dial up', 'Total']
df2[columns_to_replace] = df2[columns_to_replace].replace('[.,]', '', regex=True)

# Convertir las columnas numéricas a tipo float
df2[columns_to_replace] = df2[columns_to_replace].astype(float)

# Crear paleta de colores claros
colors = ['#86C7B8', '#F8B195', '#A8E6CE', '#F67280', '#C06C84', '#6C5B7B']

# Crear gráfico por periodo
plt.figure(figsize=(12, 6))
sns.barplot(data=df2, x='Periodo', y='Banda ancha fija', hue='Trimestre', palette=colors)
plt.title('Banda ancha fija por Trimestre')
plt.xlabel('Periodo')
plt.ylabel('Banda ancha fija')
plt.legend(title='Trimestre')
plt.xticks(rotation=45)
plt.show()

# Crear gráfico por trimestre
plt.figure(figsize=(12, 6))
sns.barplot(data=df2, x='Trimestre', y='Banda ancha fija', hue='Periodo', palette=colors)
plt.title('Banda ancha fija por Periodo')
plt.xlabel('Trimestre')
plt.ylabel('Banda ancha fija')
plt.legend(title='Periodo')
plt.xticks(rotation=45)
plt.show()

```

En ambos graficos podemos observar un gran avance durante los años. Tambien se puede notar que por trimestre y por año siempre aumenta la cifra de conexiones por banda ancha fija.

```python
# Leer el archivo CSV
archivo2 = 'C:/Users/Usuario/Downloads/Internet_BAF.csv'
df2 = pd.read_csv(archivo2)

# Reemplazar los puntos y comas por nada en los valores de las columnas numéricas
columns_to_replace = ['Dial up', 'Total']
df2[columns_to_replace] = df2[columns_to_replace].replace('[.,]', '', regex=True)

# Convertir las columnas numéricas a tipo float
df2[columns_to_replace] = df2[columns_to_replace].astype(float)

# Crear paleta de colores claros
colors = ['#86C7B8', '#F8B195', '#A8E6CE', '#F67280', '#C06C84', '#6C5B7B']

# Crear gráfico por periodo
plt.figure(figsize=(12, 6))
sns.barplot(data=df2, x='Periodo', y='Dial up', hue='Trimestre', palette=colors)
plt.title('Dial up por Trimestre')
plt.xlabel('Periodo')
plt.ylabel('Dial up')
plt.legend(title='Trimestre')
plt.xticks(rotation=45)
plt.show()

# Crear gráfico por trimestre
plt.figure(figsize=(12, 6))
sns.barplot(data=df2, x='Trimestre', y='Dial up', hue='Periodo', palette=colors)
plt.title('Dial up por Periodo')
plt.xlabel('Trimestre')
plt.ylabel('Dial up')
plt.legend(title='Periodo')
plt.xticks(rotation=45)
plt.show()
```

Estos dos graficos muestran un caso contrario a los dos anteriores, debido a que pasa el tiempo y hay una gran caida de las conexiones por dial up

```python
# Leer el archivo CSV
archivo3 = 'C:/Users/Usuario/Downloads/Internet_Accesos-por-tecnologia.csv'
df3 = pd.read_csv(archivo3)

# Reemplazar los puntos de los valores numéricos
columns_to_replace = ['ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros', 'Total']
df3[columns_to_replace] = df3[columns_to_replace].replace('\.', '', regex=True)

# Convertir las columnas numéricas a tipo float
df3[columns_to_replace] = df3[columns_to_replace].astype(float)

# Crear gráfico de líneas para las tecnologías de acceso
plt.figure(figsize=(12, 6))
plt.plot(df3['Periodo'], df3['ADSL'], label='ADSL')
plt.plot(df3['Periodo'], df3['Cablemodem'], label='Cablemodem')
plt.plot(df3['Periodo'], df3['Fibra óptica'], label='Fibra óptica')
plt.plot(df3['Periodo'], df3['Wireless'], label='Wireless')
plt.plot(df3['Periodo'], df3['Otros'], label='Otros')
plt.plot(df3['Periodo'], df3['Total'], label='Total')
plt.title('Accesos a Internet por Tecnología')
plt.xlabel('Período')
plt.ylabel('Cantidad de Accesos')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

Podemos observar claramente como es que al pasar del tiempo subieron considerablemente el uso del cable modem y la fibra optica. Mientras que evolucionaban las conexiones se limito el uso de ADSL

```python
# Leer el archivo CSV
archivo3 = 'C:/Users/Usuario/Downloads/Internet_Accesos-por-tecnologia.csv'
df3 = pd.read_csv(archivo3)

# Reemplazar los puntos de los valores numéricos
columns_to_replace = ['ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros', 'Total']
df3[columns_to_replace] = df3[columns_to_replace].replace('\.', '', regex=True)

# Convertir las columnas numéricas a tipo float
df3[columns_to_replace] = df3[columns_to_replace].astype(float)

# Filtrar el DataFrame solo para la tecnología "Wireless"
df_wireless = df3[df3['Wireless'] > 0]

# Crear gráfico de líneas para la tecnología "Wireless"
plt.figure(figsize=(12, 6))
plt.plot(df_wireless['Periodo'], df_wireless['Wireless'], label='Wireless')
plt.title('Accesos a Internet por Tecnología (Wireless)')
plt.xlabel('Período')
plt.ylabel('Cantidad de Accesos')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

Podemos notar una constancia entre los años 2014-2017(fines), en los finales del 2017 y principios del 2018 aumento drasticamente el uso de la conexion wireless y de ahi no paro hasta hoy en dia.

```python
# Leer el archivo CSV
archivo4 = 'C:/Users/Usuario/Downloads/historico_velocidad_internet.csv'
df4 = pd.read_csv(archivo4)

# Convertir la columna "Mbps (Media de bajada)" a tipo numérico
df4['Mbps (Media de bajada)'] = df4['Mbps (Media de bajada)'].str.replace(',', '.')
df4['Mbps (Media de bajada)'] = pd.to_numeric(df4['Mbps (Media de bajada)'], errors='coerce')

# Crear gráfico de barras por trimestre
plt.figure(figsize=(12, 6))
sns.barplot(data=df4, x='Trimestre', y='Mbps (Media de bajada)')
plt.title('Velocidad de Internet (Media de bajada) por Trimestre')
plt.xlabel('Trimestre')
plt.ylabel('Mbps (Media de bajada)')
plt.xticks(rotation=45)
plt.show()

# Crear gráfico de barras por año
plt.figure(figsize=(12, 6))
sns.barplot(data=df4, x='Año', y='Mbps (Media de bajada)')
plt.title('Velocidad de Internet (Media de bajada) por Año')
plt.xlabel('Año')
plt.ylabel('Mbps (Media de bajada)')
plt.xticks(rotation=45)
plt.show()
```

El primer grafico podemos notar una subida hasta el 3r trimestre que despues hay un decrecimiento para el 4to trimestre.

Y para el segundo se puede observar una clara evolucion de la velocidad de internet a traves de los años

```python
# Leer el archivo CSV
archivo5 = 'C:/Users/Usuario/Downloads/Internet_Ingresos.csv'
df5 = pd.read_csv(archivo5)

# Eliminar los puntos de la columna "Ingresos (miles de pesos)" y convertir a tipo numérico
df5['Ingresos (miles de pesos)'] = df5['Ingresos (miles de pesos)'].str.replace('.', '').astype(float)

# Convertir los ingresos a pesos dividiendo por mil
df5['Ingresos (miles de pesos)'] = df5['Ingresos (miles de pesos)'] * 1000

# Crear gráfico de barras para los ingresos por trimestre
plt.figure(figsize=(12, 6))
sns.barplot(data=df5, x='Trimestre', y='Ingresos (miles de pesos)')
plt.title('Ingresos de Internet por Trimestre')
plt.xlabel('Trimestre')
plt.ylabel('Ingresos (pesos)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Crear gráfico de barras para los ingresos por año
plt.figure(figsize=(12, 6))
sns.barplot(data=df5, x='Año', y='Ingresos (miles de pesos)')
plt.title('Ingresos de Internet por Año')
plt.xlabel('Año')
plt.ylabel('Ingresos (pesos)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Se puede observar lo mismo que los dos graficos de arriba, en el grafico de trimestre notamos una aumento hasta el 3er trimestre y en el 4to una caida de ingresos.
